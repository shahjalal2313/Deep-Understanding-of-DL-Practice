**Note:** 
* Log is a monotonic function of x. it means that when x goes up log also goes up.
* In Optimization of DL,monotonicity is important becuase minimizing x is the same thing as minimizing log(x)! (n.b only for x>0)
* Log is a non-linear function. 
* As the values of x increases the space between the data points gets closer
* Log stretches out the small values of x.In DL we will work with really small values like probability values, Loss Function so minimizing log is convenient than the original value.


```python
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0001,1,200)
#will return 200 evenly spaced numbers over a specified interval

#computing log of x
logx = np.log(x)
```

```python
#plotting x vs logx
fig = plt.figure(figsize=(10,4))
plt.plot(x,logx,'ks-',markerfacecolor="w")
plt.xlabel("Values of X")
plt.ylabel("$Log(x)$")
plt.title("x vs $Log(x)$")
plt.show()
```

```python
#plotting x vs logx
exp = np.exp(x)
fig = plt.figure(figsize=(10,4))
plt.plot(x,exp,'ks-',markerfacecolor="w")
plt.xlabel("Values of x")
plt.ylabel("$e^x$")
plt.title("x vs $e^x$")
plt.show()
```

**Log and exp are inverses**

```python
#experimenting with fewer points by slicing
x_s = x[:20]
logx_s = logx[:20]
exp_s = exp[:20]
```

```python
#plotting
plt.plot(x_s,x_s,color=[.8,.8,.8])
plt.plot(x_s,np.exp(logx_s),'o',markersize=7,color=[.1,.2,.8])
plt.plot(x_s,np.log(exp_s),'x',markersize=9,color=[.8,.2,.1])
plt.xlabel('x')
plt.ylabel('f(g(x))')
plt.legend(["unity",'exp(log(x))','log(exp(x))'])
plt.show()
```

**Logerithm and natural exponential is reverse to each other:** *From the graph we can see that 2nd and 3rd graph cancels each other and form a straigt line like the first one which is not a mathmatical proof but a illustration of the fact*
